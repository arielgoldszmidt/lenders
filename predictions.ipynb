## Make regression tables for LaTeX

```python
def to_table(regs, xnames):
    table_body = ""

    for var in xnames:
        table_body += var
        for i in range(8):
            table_body += " & " + '${0:.3f}'.format(regs[i].params[var])
            p = regs[i].pvalues[var]
            if p < 0.005:
                table_body += "^{***}"
            elif p < 0.01:
                table_body += "^{**}"
            elif p < 0.05:
                table_body += "^{*}"
            table_body += "$"
        table_body += " \\\\\n "
        for i in range(8):
            table_body += " & $(" + '{0:.3f}'.format(regs[i].HC0_se[var]) + ")$"
        table_body += " \\\\\n "
        table_body += " & " * 8
        table_body += " \\\\\n "

    table_body += "\\hline \\\\[-1.8ex] \n"
    table_body += "$R^2$"
    for i in range(8):
        table_body += " & " + '${0:.3f}$'.format(regs[i].rsquared)
    table_body += " \\\\\n "

    table_body += "Adj. $R^2$"
    for i in range(8):
        table_body += " & " + '${0:.3f}$'.format(regs[i].rsquared_adj)
    table_body += " \\\\\n "

    table_body += "$N$"
    for i in range(8):
        table_body += " & " + '${0:.0f}$'.format(regs[i].nobs)    

    print(table_body.replace('_', '\_'))
    
    return
```

## Logit table

```python
def to_table_logit(regs, xnames):
    table_body = ""

    for var in xnames:
        table_body += var
        for i in range(8):
            table_body += " & " + '${0:.3f}'.format(regs[i].params[var] * 100)
            p = regs[i].pvalues[var]
            if p < 0.005:
                table_body += "^{***}"
            elif p < 0.01:
                table_body += "^{**}"
            elif p < 0.05:
                table_body += "^{*}"
            table_body += "$"
        table_body += " \\\\\n "
        for i in range(8):
            table_body += " & $(" + '{0:.3f}'.format(regs[i].bse[var] * 100) + ")$"
        table_body += " \\\\\n "
        table_body += " & " * 8
        table_body += " \\\\\n "

    table_body += "\\hline \\\\[-1.8ex] \n"
    table_body += "$R^2_{\\mathrm{McFadden}}$"
    for i in range(8):
        table_body += " & " + '${0:.3f}$'.format(regs[i].prsquared)
    table_body += " \\\\\n "

    table_body += "$-2\\log(\\mathrm{LR})$"
    for i in range(8):
        table_body += " & " + '${0:.3f}$'.format(regs[i].llr)
    table_body += " \\\\\n "

    table_body += "$N$"
    for i in range(8):
        table_body += " & " + '${0:.0f}$'.format(regs[i].nobs)    

    print(table_body.replace('_', '\_'))
    
    return
```
